# Fazendo importações da biblioteca

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Carregando os dados do arquivo CSV

```python
data = pd.read_csv('map-o.1.csv')
```

# Gráfico de Linhas
## Análise das tendências semanais de cada LGA 


```python
# Análise das tendências semanais de cada LGA
# Gráfico de Linhas
plt.figure(figsize=(10, 6))
for lga in data['lga_code'].unique():
    lga_data = data[data['lga_code'] == lga]
    plt.plot(lga_data.columns[2:], lga_data.values[0][2:], label=f'LGA {lga}')
plt.xlabel('Semana')
plt.ylabel('Valor')
plt.title('Tendências Semanais por LGA')
plt.legend()
plt.show()
```

# Gráfico de Barras
##  Análise da distribuição dos valores em cada semana 


```python
# Agrupando os dados por 'lga_code' e calculando a média para cada semana
data_grouped = data.groupby('lga_code').mean()

# Gráfico de Barras
plt.figure(figsize=(10, 6))
for week in data_grouped.columns:
    plt.bar(data_grouped.index, data_grouped[week], label=week)
plt.xlabel('LGA Code')
plt.ylabel('Valor Médio')
plt.title('Distribuição de Valores Médios por Semana')
plt.legend()
plt.show()
```

# Histograma
## Análise da frequência de valores em uma semana específica 


```python
# Análise da frequência de valores em uma semana específica
# Histograma
week_to_analyze = 'week_3'
plt.figure(figsize=(10, 6))
plt.hist(data[week_to_analyze], bins=10)
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.title(f'Histograma de Valores para a Semana {week_to_analyze}')
plt.show()
```
